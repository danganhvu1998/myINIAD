### 3D plot of function  

#### Import libraries  

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import cm
from mpl_toolkits.mplot3d import axes3d
```

```python
%config InlineBackend.figure_formats = {'png', 'retina'}
```

#### Ackley function 2D (modified)  

```python
def ackley2d(x, y):
    total = 20.0
    s1 = (x**2 + y**2)/2
    total -= 20.0 * np.exp(-0.2 * s1**0.5)
    s2 = np.cos(2*np.pi*x) + np.cos(2*np.pi*y)
    total -= np.exp(s2 / 2)
    return total
```

#### 3D plot  

```python
fig = plt.figure()

ax = fig.add_subplot(111, projection='3d')
x = np.linspace(-32, 32, 50)
y = np.linspace(-32, 32, 50)
X, Y = np.meshgrid(x, y)
Z = ackley2d(X, Y)
plt.title("Ackley function")
plt.xlabel("x")
plt.ylabel("y")
ax.plot_surface(X, Y, Z, cmap=cm.coolwarm, linewidth=0, rstride=1, cstride=1, alpha=.5)
ax.contour(X, Y, Z, 10, cmap=cm.coolwarm, linestyles="solid", offset=-1)
plt.show()
```
