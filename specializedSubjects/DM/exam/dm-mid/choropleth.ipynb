## Example of displaying heatmap using folium (English)  

### Import libraries  

```python
import pandas as pd
import folium
import json
```

### Parameters

```python
csv_in = 'dm-mid-p1-1.csv'
csv_1_in = 'dm-mid-p1-2.csv'
geo = 'japan.json'
pd.options.display.max_rows = 999
pd.options.display.max_columns = 999
```

### Read CSV file

```python
df1 = pd.read_csv(csv_in, delimiter=',', skiprows=0, header=0)
# print(df['Year'] == 2016)
print(df1.shape)
print(df1.info())
display(df1.head())
#display(df)
```

```python
df2 = pd.read_csv(csv_1_in, delimiter=',', skiprows=0, header=0)
# print(df['Year'] == 2016)
print(df2.shape)
print(df2.info())
display(df2.head())
#display(df)
df12 = pd.merge(df1, df2, left_on='prefecture', right_on='pref', how='inner')
df12['pop/elem_total'] = df12['pop'] / df12['elem_total']
display(df12.head())
```

```python
m = folium.Map(location=[38,135], zoom_start=5)
folium.Choropleth(
    geo_data=geo,
    key_on='feature.properties.name',
    data=df1,
    columns=['prefecture', 'elem_male'],
    fill_color='BuPu',
    legend_name='elem_male'
).add_to(m)
m.save('elem_male.html')
display(m)
```

```python
m = folium.Map(location=[38,135], zoom_start=5)
folium.Choropleth(
    geo_data=geo,
    key_on='feature.properties.name',
    data=df12,
    columns=['prefecture', 'pop/elem_total'],
    fill_color='BuPu',
    legend_name='pop/elem_total'
).add_to(m)
m.save('pop_elem_total.html')
display(m)
```
