```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from pandas.plotting import register_matplotlib_converters
from sklearn import preprocessing
register_matplotlib_converters()
```

```python
csv_android = './android-data.csv'
csv_ios = './ios-data.csv'
```

```python
def time_data_frame_to_graphs(df, data_cols = ["app_average_cpu", "app_average_ram", "app_average_translate_times", "other_data_that_we_might_want_to_collect"], time_col = "time"):
    """
    df: pandas dataframe
    data_cols: list of columns that we want to plot
    time_col: column that we want to use as x axis
    """

    # Graphs by col
    for col in data_cols:
        plt.plot(df[time_col], df[col])
        plt.xlabel('Month')
        plt.ylabel(col)
        plt.show()
    
    # General graph
    for col in data_cols:
        # first, we have to normalize the data to be between 0 and 1
        normalized_col =  f'{col}_normalized'
        df[normalized_col] = (df[col] - df[col].mean()) / df[col].std()
        # then draw the graph
        plt.plot(df[time_col], df[normalized_col])
    plt.legend(data_cols)
    plt.show()
```
