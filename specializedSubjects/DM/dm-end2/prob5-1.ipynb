### Test of dual_annealing  

#### Import libraries  

```python
import numpy as np
import pandas as pd
from random import uniform
from scipy.optimize import dual_annealing
import matplotlib.pyplot as plt
```

#### Cost function  
Known as Ackley function  

```python
def ackley2d(x):
    X = x[0]
    Y = x[1]
    total = X**2 + Y **2 - 0.5 * X - Y + np.cos(10*(X-Y))
    # or
    # total = 20.0
    # s1 = np.sum(x**2) / 2
    # total -= 20.0 * np.exp(-0.2 * s1**0.5)
    # s2 = np.sum(np.cos(2*np.pi*x))
    # total -= np.exp(s2 / 2)
    return total
```

#### Set bounds (lower, upper) for each x[i]  

```python
bounds = [(-3.0, 3.0), (-3.0, 3.0)]
```

#### Multiple trial and check obtained minima  

```python
n_trial = 50
```

```python
x = np.zeros(2)
for i in range(n_trial):
    # Initial value
    x[0] = uniform(-3.0, 3.0)
    x[1] = uniform(-3.0, 3.0)
    print(x)  # debug
    
    # Dual annealing optimization
    ret = dual_annealing(ackley2d, bounds, x0=x, maxiter=500, seed=19)
    print('x:', ret.x)
    print('f(x):', ret.fun)
```
