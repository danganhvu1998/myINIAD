### Import libraries  

```python
import pandas as pd
import folium
```

### Parameters  

```python
csv_in = 'swimming_pool.csv'
geo = 'japan.json'
```

### Read CSV files  

```python
df = pd.read_csv(csv_in, delimiter=',', skiprows=5, header=0)
print(df.shape)
print(df.info())
display(df.head())
```

```python
m = folium.Map(location=[38,135], zoom_start=5)
folium.Choropleth(
    geo_data=geo,
    key_on='feature.properties.name',
    data=df,
    columns=['Pref', 'Swimming_Pool'],
    fill_color='BuPu',
    legend_name='#SP'
).add_to(m)

display(m)
```

```python
df['pool_per_1m_pepple'] = df['Swimming_Pool']/ df['Population(M)']
m = folium.Map(location=[38,135], zoom_start=5)
folium.Choropleth(
    geo_data=geo,
    key_on='feature.properties.name',
    data=df,
    columns=['Pref', 'pool_per_1m_pepple'],
    fill_color='BuPu',
    legend_name='#SP/PP'
).add_to(m)

display(m)
```
