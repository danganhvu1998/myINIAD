```python
import numpy as np
import matplotlib.pyplot as plt
from random import uniform
from scipy.optimize import dual_annealing
```

```python
def f(X):
    y = X[0]
    x = X[1]
    # return X + Y
    return 2*(x-y)**2 + 2 * np.sin(x+y) -3*x + 5*y + 6
```

```python
bounds = [(-1.5, 4.0), (-3.0, 4.0)]
n_trial = 2
```

```python
X = []
Y = []
for i in np.arange(bounds[0][0],bounds[0][1], 0.3):
    for j in np.arange(bounds[1][0],bounds[1][1], 0.3):
        X.append(i)
        Y.append(j)
X = np.array(X)
Y = np.array(Y)
x = np.array([X, Y])
z = f(x)
# print(X)
# print(Y)
# print(z)

# Creating figure
fig = plt.figure(figsize = (10, 7))
ax = plt.axes(projection ="3d")
 
# Creating plot
ax.scatter3D(X, Y, z, color = "green")
plt.title("simple 3D scatter plot")
 
# show plot
plt.show()
```

```python
x = np.zeros(2)
for i in range(n_trial):
    # Initial value
    x[0] = uniform(bounds[0][0],bounds[0][1])
    x[1] = uniform(bounds[1][0],bounds[1][1])
    print(x)  # debug
    
    # Dual annealing optimization
    ret = dual_annealing(f, bounds, x0=x, maxiter=500, seed=19)
    print('x:', ret.x)
    print('f(x):', ret.fun)
```
