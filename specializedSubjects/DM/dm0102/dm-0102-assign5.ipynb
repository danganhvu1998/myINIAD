## Example of displaying heatmap using folium (English)  

### Import libraries  

```python
import pandas as pd
import folium
import json
```

### Parameters

```python
csv_in = '../iniadMaterial/2021/dm-0102/crime2006_2016_en-utf8.csv'
geo = '../iniadMaterial/2021/dm-0102/japan.json'
pd.options.display.max_rows = 999
pd.options.display.max_columns = 999
```

### Read CSV file

```python
df = pd.read_csv(csv_in, delimiter=',', skiprows=0, header=0)
# print(df['Year'] == 2016)
df = df[(df['Year'] == 2016)]
print(df.shape)
print(df.info())
display(df.head())
#display(df)
```

```python
m = folium.Map(location=[38,135], zoom_start=5)
folium.Choropleth(
    geo_data=geo,
    key_on='feature.properties.name',
    data=df,
    columns=['Pref', '#Arrested'],
    fill_color='BuPu',
    legend_name='Arrested'
).add_to(m)

display(m)
```
