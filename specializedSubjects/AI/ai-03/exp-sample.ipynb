### plot of exp function

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(-2, 2)
plt.plot(x, np.exp(-x), label='exp(-x)')
plt.grid(True)
plt.legend()
plt.show()
```
