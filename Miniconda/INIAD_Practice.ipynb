```python
import numpy as np
import matplotlib.pyplot as plt
import timeit
```

```python
x = np.linspace(1,10,100)
y1 = x
y2 = np.log(x)
y3 = x*np.log(x)
plt.plot(x,y1,color="red",label='$y=x$')
plt.plot(x,y2,color="blue",label='$y=\log x$')
plt.plot(x,y3,color="green",label='$y=x \log x$')
plt.xlim([1,10])
plt.legend()
plt.xlabel('x',fontsize=9)
plt.ylabel('y',fontsize=9)
plt.show()
```

```python
def f(x):
    s = 0
    for i in range(x):
        for j in range(x):
            s+=1
    return s

def ff(x):
    s = 0
    for i in range(x):
        s+=1
    return s

def fff(x):
    s = 0
    y = int(np.sqrt(x))
    for i in range(x):
        for j in range(y):
            s+=1
    return s
```
