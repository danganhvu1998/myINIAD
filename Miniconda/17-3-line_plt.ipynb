```python
import numpy as np
import matplotlib.pyplot as plt
elec=np.loadtxt('data/elec.csv', delimiter=',', skiprows=1)
```

```python
print( elec.shape )
print( elec )
```

```python
mon=elec[1::2, 0]
e1=elec[1::2, 1]
e2=elec[1::2, 2]
```

```python
plt.plot(mon, e1, label='2008')
plt.plot(mon, e2, label='2015')
plt.xlabel('Month')
plt.ylabel('Yen')
plt.legend()
plt.show()
```
