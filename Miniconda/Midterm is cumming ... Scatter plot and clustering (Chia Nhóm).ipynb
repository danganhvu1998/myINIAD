```python
#Import
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
```

```python
iris_sep=np.loadtxt('data/iris_sepal.csv', delimiter=',', skiprows=1) #how to read the file
print( iris_sep.shape ) #150x2 arrya
```

```python
x=iris_sep[:, 0] #first collum
y=iris_sep[:, 1] #second collum
print( x.shape )
print( y.shape )
```

```python
plt.scatter(x, y, marker='o', label='iris_sepal')
plt.xlabel('Length [cm]')
plt.ylabel('Width [cm]')
plt.legend()
plt.show()
```

```python
kmcls=KMeans(n_clusters=2).fit_predict(iris_sep)
print(kmcls)
x0=x[kmcls==0]
y0=y[kmcls==0]
x1=x[kmcls==1]
y1=y[kmcls==1]
plt.scatter(x0, y0, marker='o', c='blue', label='cluster0')
plt.scatter(x1, y1, marker='D', c='red', label='cluster1')
plt.xlabel('Length [cm]')
plt.ylabel('Width [cm]')
plt.legend()
plt.show()

#Down there just change the appearance
plt.scatter(x0, y0, marker='$0$', c='blue', label='cluster0')
plt.scatter(x1, y1, marker='$1$', c='red', label='cluster1')
plt.xlabel('Length [cm]', fontsize=20)
plt.ylabel('Width [cm]', fontsize=20)
plt.tick_params(labelsize=20)
plt.xlim([2, 10])
plt.ylim([1, 5.5])
plt.title('Results of KMeans(2)', fontsize=20)
plt.legend()
plt.show()
```

```python
kmcls=KMeans(n_clusters=3).fit_predict(iris_sep)
print(kmcls)
x0=x[kmcls==0]
y0=y[kmcls==0]
x1=x[kmcls==1]
y1=y[kmcls==1]
x2=x[kmcls==2]
y2=y[kmcls==2]
plt.scatter(x0, y0, marker='o', c='blue', label='cluster0')
plt.scatter(x1, y1, marker='D', c='red', label='cluster1')
plt.scatter(x2, y2, marker='x', c='green', label='cluster2')
plt.xlabel('Length [cm]')
plt.ylabel('Width [cm]')
plt.legend()
plt.show()
```
