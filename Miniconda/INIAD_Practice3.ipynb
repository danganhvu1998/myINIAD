```python
import timeit
import matplotlib.pyplot as plt
import random
```

```python
def ssort(lst):
    for i in range(len(lst)-1):
        min = lst[i]
        minpos = i
        for j in range(i+1, len(lst)):
            if lst[j]<min:
                min = lst[j]
                minpos = j
        if i!=minpos:
            tmp = lst[i]
            lst[i] = min
            lst[minpos] = tmp
    return lst
```

```python
def qsort(lst):
    if len(lst)<=1:
        return lst
    pivot = lst[0]
    lst1 = [item for item in lst if item < pivot]
    lst2 = [item for item in lst if item > pivot]
    return qsort(lst1) + [pivot] + qsort(lst2)
```

```python
def setup_lst(x):
    lst = random.sample(range(x*2),k=x)
    return lst
```

```python
n = 1
r = 100
```

```python
xs = range(20,500,20)
```

```python
ys0 = []
for x in xs:
    time = timeit.repeat('ssort(lst)',setup='lst = setup_lst(x)', globals = globals(), repeat = r, number = n)
    ys0.append(sum(time)/(n*r))
```

```python
plt.plot(xs,ys0,'o')
plt.ylim(0,max(ys0)*1.1)
plt.xlabel('Length of list',fontsize=9)
plt.ylabel('Average time (s)',fontsize=9)
plt.show()
```

```python
ys1 = []
for x in xs:
    time = timeit.repeat('qsort(lst)',setup='lst = setup_lst(x)', globals = globals(), repeat = r, number = n)
    ys1.append(sum(time)/(n*r))
```

```python
plt.plot(xs,ys1,'d')
plt.ylim(0,max(ys1)*1.1)
plt.xlabel('Length of list',fontsize=9)
plt.ylabel('Average time (s)',fontsize=9)
plt.show()
```

```python
ys2 = []
for x in xs:
    time = timeit.repeat('qsort(lst)',setup='lst = range(x)', globals = globals(), repeat = r, number = n)
    ys2.append(sum(time)/(n*r))
```

```python
plt.plot(xs,ys1,'d')
plt.plot(xs,ys0,'o')
plt.plot(xs,ys2,'x')
plt.ylim(0,max(ys2)*1.1)
plt.xlabel('Length of list',fontsize=9)
plt.ylabel('Average time (s)',fontsize=9)
plt.show()
```
