```python
#Import
import numpy as np
import matplotlib.pyplot as plt
from keras.models import Sequential
from keras.layers import Dense, Activation
```

```python
#Load data
#dat=np.loadtxt('data/points_and.csv', delimiter=',', skiprows=1) #Default Data
dat=np.loadtxt('data/points_MyData.csv', delimiter=',', skiprows=1) #My data
data_train=dat[:,:-1] #First 2 collumn
class_train=dat[:,-1] #Last collumn
print('data:', data_train.shape)
print('class:', class_train.shape)
```

```python
#Training
model=Sequential()
model.add(Dense(10,input_dim=2,activation='relu'))
model.add(Dense(10,activation='relu'))
model.add(Dense(1,activation='sigmoid'))
model.compile(optimizer='rmsprop',loss='binary_crossentropy',metrics=['accuracy'])
print(model.summary())
fit_log = model.fit(data_train,class_train,epochs=2000,batch_size=1)
#model=Sequential()
#model.add(Dense(1,input_dim=2,activation='sigmoid'))
#model.compile(optimizer='sgd',loss='binary_crossentropy',metrics=['accuracy'])
#fit_log=model.fit(data_train,class_train,epochs=1000,batch_size=1)
```

```python
#Init Graph
plt.plot(fit_log.history['acc'])
plt.title('Model Accuracy')
plt.ylabel('Accuracy')
plt.xlabel('Epoch')
plt.legend(['train'], loc='best')
plt.show()
```

```python
#What the hell is this?
pred=model.predict(data_train)
print(pred)
```

```python
#Take value of dataset to draw 21x21 result graph
t1=data_train[class_train==1]
t0=data_train[class_train==0]
t1_x=t1[:,0]
t1_y=t1[:,1]
t0_x=t0[:,0]
t0_y=t0[:,1]
```

```python
#Predict value of dataset to draw 21x21 result graph
g=np.loadtxt('data/grid01_21x21.csv', delimiter=',')
pred_g=model.predict(g)[:, 0]
g1=g[pred_g>=0.5]
g0=g[pred_g<0.5]
g1_x=g1[:, 0]
g1_y=g1[:, 1]
g0_x=g0[:, 0]
g0_y=g0[:, 1]
```

```python
plt.scatter(t1_x,t1_y,marker='s',facecolors='none',edgecolors='red',s=180)
plt.scatter(t0_x,t0_y,marker='o',facecolors='none',edgecolors='red',s=180)
plt.scatter(g1_x,g1_y,marker='$1$')
plt.scatter(g0_x,g0_y,marker='$0$')
plt.gca().set_aspect('equal', adjustable='box')
plt.xlim(-4.5,4.5)
plt.ylim(-4.5,4.5)
plt.xlabel('X')
plt.ylabel('Y')
plt.show()
```
